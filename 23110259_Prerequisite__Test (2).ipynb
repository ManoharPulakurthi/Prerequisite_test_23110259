

   Given,

Two matrices A and B of size \\(n*k\\) and \\(k*m\\) respectively.

Therefore if we multiply two matrices the resultant matrix is of size \\(n*m\\).

dot product of ith row of A and jth column of B is performed k times.

so the number of multiplications are \\(n*m*k\\).

in dot product of ith row of A and jth column of B the number of additions are k-1, because adding k numbers needs only kâˆ’1 additions.

Hence total number of additions is \\(n*m*(k-1)\\).



```python
#2

import numpy as np
import time

a = np.array([
    [1, 2, 3],
    [4, 5, 6]])
b = np.array([
    [7, 8],
    [9, 10],
    [11, 12]])

start_list = time.time()
l = []
if len(a[0]) != len(b):
    print('Not Possible')
else:
    for i in range(len(a)):
        k = []
        for j in range(len(b[0])):
            s = 0
            for m in range(len(b)):
                s += a[i][m] * b[m][j]
            k.append(s)
        l.append(k)
    result_list = np.array(l)
    print(result_list)
end_list = time.time()

A = np.array([
    [1, 2, 3],
    [4, 5, 6]])
B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]])

start_numpy = time.time()
C = np.dot(A, B)
print(C)
end_numpy = time.time()

print("List time:", end_list - start_list)
print("NumPy dot time:", end_numpy - start_numpy)
```

The Python list of lists approach uses loops that are interpreted one by one, which is slow. whereas numpy is faster because to perform matrix multiplications it uses C language for efficient calculations and also it performs the calculations of all the arrays at once.

\\[f(x) = x^2y + y^3\sin(x)\\]

Gradient of the function with respect to x is:

\\[\frac{d(f(x,y))}{dx} =\\]

\\[\frac{d(x^2y)}{dx} + \frac{d(y^3\sin(x))}{dx}\\]

\\[\frac{d(f(x,y))}{dx} = 2xy + y^3\cos(x)\\]

Gradient of the function with respect to y is:

\\[\frac{d(f(x,y))}{dy} =\\]

\\[\frac{d(x^2y)}{dy} + \frac{d(y^3\sin(x))}{dy}\\]

\\[\frac{d(f(x,y))}{dy} = x^2 + 3y^2\sin(x)\\]

```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    def __repr__(self):
        return f"{self.name} ({self.roll_number}): {self.marks}"


class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def __repr__(self):
        return f"Branch: {self.name}, Students: {len(self.students)}"


class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

    def __repr__(self):
        return f"Year: {self.year}, Branches: {len(self.branches)}"


class Database:
    def __init__(self):
        self.years = []

    def add_year(self, year):
        self.years.append(year)

    def __repr__(self):
        return f"Database with {len(self.years)} year(s)"
```

```python
#9
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0.5,100.5,0.5)
y1 = x
y2 = x**2
y3 = (x**3)/100
y4 = np.sin(x)
y5 = (np.sin(x))/x
y6 = np.log(x)
y7 = np.exp(x)
fig = plt.figure(figsize=(12, 8))
plt.subplot(3,3,1)
plt.plot(x,y1)
plt.title('y = x')
plt.subplot(3,3,2)
plt.plot(x,y2)
plt.title('y = x**2')
plt.subplot(3,3,3)
plt.plot(x,y3)
plt.title('y = x**3/100')
plt.subplot(3,3,4)
plt.plot(x,y4)
plt.title('y = sin(x)')
plt.subplot(3,3,5)
plt.plot(x,y5)
plt.title('y = sin(x)/x')
plt.subplot(3,3,6)
plt.plot(x,y6)
plt.title('y = log(x)')
plt.subplot(3,3,7)
plt.plot(x,y7)
plt.title('y = exp(x)')
plt.tight_layout()
```

```python
#10
import numpy as np
import pandas as pd

data = np.random.rand(20, 5) + 1
df = pd.DataFrame(data, columns=list("abcde"))

std_series = df.std(ddof=0)
max_std_col = std_series.idxmax()
max_std_val = std_series.max()

row_means = df.mean(axis=1)
min_mean_index = row_means.idxmin()
min_mean_value = row_means.min()

print("Generated DataFrame:\n", df)
print(f"\nColumn with highest standard deviation: '{max_std_col}' ({max_std_val})")
print(f"Row with lowest mean: index {min_mean_index} (mean = {min_mean_value})")
```
